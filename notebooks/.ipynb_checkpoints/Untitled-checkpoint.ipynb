```python
from bs4 import BeautifulSoup
import pandas as pd
import re
from tqdm.notebook import tqdm

# Make sure tqdm shows progress bar
tqdm.pandas()

def parse_html(html):
    """Extract title, main text, and word count from HTML."""
    try:
        soup = BeautifulSoup(html, "html.parser")

        # Extract title
        title = soup.title.string.strip() if soup.title else ""

        # Extract main content text
        main_content = []
        for tag in soup.find_all(["p", "article", "main"]):
            text = tag.get_text(separator=" ", strip=True)
            if text:
                main_content.append(text)

        body_text = " ".join(main_content)
        body_text = re.sub(r"\s+", " ", body_text).strip()

        word_count = len(body_text.split())

        return pd.Series([title, body_text, word_count])

    except Exception as e:
        return pd.Series(["", "", 0])

# Apply to dataset
df[["title", "body_text", "word_count"]] = df["html_content"].apply(parse_html)

# Save parsed content
output_path = "seo-content-detector/data/extracted_content.csv"
df.to_csv(output_path, index=False)

print("✅ Parsed and saved to:", output_path)
df.head(10)
```

```python
# --- STEP 2: Text Preprocessing & Feature Engineering (TF-IDF version) ---

import pandas as pd
import numpy as np
import re
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from textstat import flesch_reading_ease
from nltk.corpus import stopwords

# download stopwords if needed
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load parsed dataset
df = pd.read_csv("seo-content-detector/data/extracted_content.csv")

# --- 1️⃣ Clean text ---
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^a-z0-9\s]', '', text)
    return text.strip()

df["clean_text"] = df["body_text"].apply(clean_text)

# --- 2️⃣ Basic metrics ---
df["sentence_count"] = df["clean_text"].apply(lambda x: len(re.split(r"[.!?]", x)))
# word_count already exists

# --- 3️⃣ Readability ---
df["flesch_reading_ease"] = df["clean_text"].apply(
    lambda x: flesch_reading_ease(x) if len(x.split()) > 0 else 0
)

# --- 4️⃣ TF-IDF keywords + vector representation ---
vectorizer = TfidfVectorizer(stop_words="english", max_features=2000)
tfidf_matrix = vectorizer.fit_transform(df["clean_text"])
feature_names = np.array(vectorizer.get_feature_names_out())

def top_keywords(text, n=5):
    try:
        response = vectorizer.transform([text])
        scores = response.toarray().flatten()
        top_indices = scores.argsort()[::-1][:n]
        return "|".join(feature_names[top_indices])
    except:
        return ""

df["top_keywords"] = df["clean_text"].apply(top_keywords)

# store the TF-IDF vector itself (as list) for later similarity
df["embedding"] = list(tfidf_matrix.toarray())

# --- 5️⃣ Save in required format ---
final_features = df[[
    "url", "word_count", "sentence_count",
    "flesch_reading_ease", "top_keywords", "embedding"
]]

output_path = "seo-content-detector/data/features.csv"
final_features.to_csv(output_path, index=False)
print("✅ Features extracted (TF-IDF only) and saved to:", output_path)
final_features.head(10)
```
